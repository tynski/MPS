```python
import numpy as np
```

```python
opady_prn = open('opady.prn', 'r')
records  = opady_prn.readlines()
len(records)
```

```python
opady = []
for x in records:
    value = ''
    for i in range(len(x) - 2, 0,-1):
        if x[i] == ' ':
            break
        value += x[i]
    record = float(value[::-1]) #reverse string and cast into float
    opady.append(record)
```
