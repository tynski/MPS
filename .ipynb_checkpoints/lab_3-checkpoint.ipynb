```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy.constants import Stefan_Boltzmann, convert_temperature
from scipy.optimize import *
```

```python
### Mean Temp. W/O Atmosphere ###

######## input data

A = 0.3     # Mean Albedo
S = 1366    # Solar Constant
stefan_boltzmann = Stefan_Boltzmann    # provided by scipy library

######## Calculation of mean temperature of earth without greenhouse effect

T_earth_surface = pow((((1.0-A) *S) / (4*stefan_boltzmann)),(1/4)) # kelvin

T_earth_surface_c = convert_temperature(val=T_earth_surface,new_scale='Celsius',old_scale='Kelvin')

print("Mean Temperature of Earth without Atmosphere: ",T_earth_surface_c," celsius")
```

```python
### Energy Balance Equations (NO Atmosphere)

area_earth = 4 * np.pi * pow((6378),2)  # reference Wikipedia Earth Radius

P_sl = S * (area_earth / 4) * (1 - A)  # Short Wave Radiation

P_z = P_sl   # Long Wave Radiation

print("Long and Short Wave Radiation : ",P_z,"\n")

### inputs for Non-Linear Equation ###

### short wawe radiation

a_s = 0.11
t_a = 0.53
a_a = 0.30

### Long wawe radiation

t_aH = 0.06
a_aH = 0.31

#####
c = 2.7



###### Solving the non-linear equation of energy Balance Equation (WITH Atmosphere)

def equations_v2(z):
    T_a = z[0]
    T_s = z[1]
    
    F = np.empty((2))
    
    F[0] = ((- t_a)*(1 - a_s)*(S/4) +
    c*(T_s - T_a) +
    stefan_boltzmann*(pow(T_s,4))*(1 - a_aH) -
    stefan_boltzmann*(pow(T_a,4))) 
    
    F[1] = (- (1 - a_a - t_a + a_s*t_a)*(S/4) -
    c*(T_s - T_a) -
    stefan_boltzmann*(pow(T_s,4))*(1 - t_aH - a_aH) +
    2*stefan_boltzmann*(pow(T_a,4)))
    
    return F


guess = np.array([0,0])

z = fsolve(equations_v2,guess)


print("Mean Temperature of the Atmosphere : ",convert_temperature(val=z[0],new_scale='Celsius',old_scale='Kelvin')," 'C" )
print("Mean Surface Temperature : ",convert_temperature(val=z[1],new_scale='Celsius',old_scale='Kelvin')," 'C" )
```

```python
######### Visualization to Compare ###########

S_range = S # duplicated for loop usage

# array filled with results
Ta_store = []
Ts_store = []
S_store = []

# filling the empty arrays
for i in range(int(S_range*0.8),int(S_range*1.2)):    
    S = i
    z = fsolve(equations_v2,guess)
    Ta_store.append(convert_temperature(val=z[0],new_scale='Celsius',old_scale='Kelvin'))    
    Ts_store.append(convert_temperature(val=z[1],new_scale='Celsius',old_scale='Kelvin'))    
    S_store.append(i)
    
# drawing a Plot Graph to see the results    
plt.figure(figsize=(14,6))
plt.suptitle("Mean Temperatures / Solar Constant(0.8S-1.2S)",fontsize=16)
plt.subplot(1,2,1)
plt.title("Mean Temp. of the Atmosphere")
plt.plot(Ta_store,S_store)
plt.grid()
plt.xlabel("Temperature (Celsius)")
plt.ylabel("Solar Constant")
plt.subplot(1,2,2)
plt.title("Mean Surface Temp.")
plt.plot(Ts_store,S_store)
plt.grid()
plt.xlabel("Temperature (Celsius)")
plt.ylabel("Solar Constant")
plt.show()

print("As the Solar Constant increase, Mean tempratures are increasing too")
```

```python
S = S_range  # restoring S for finding critical points

# To find Corresponding S value for -10 celsius mean Surface Temp we 
# will use a bigger gap of possible S Values

# array filled with results
Ta_store = []
Ts_store = []
S_store = []

# filling the empty arrays
for i in range(int(S_range*0.5),int(S_range*1.5)):    
    S = i
    z = fsolve(equations_v2,guess)
    Ta_store.append(convert_temperature(val=z[0],new_scale='Celsius',old_scale='Kelvin'))    
    Ts_store.append(convert_temperature(val=z[1],new_scale='Celsius',old_scale='Kelvin'))    
    S_store.append(i)
```

```python
plt.figure(figsize=(12,8))
plt.suptitle("Mean Temperatures / Solar Constant(0.5S-1.5S)",fontsize=16)
plt.plot(Ts_store,S_store)
plt.xlabel("Temperature (Celsius)")
plt.ylabel("Solar Constant")
plt.grid()
plt.show()

for i in range(500):
    if(Ts_store[i] > -10.1 and Ts_store[i] < -9.9):
        print("Critical Solar Constant for -10 Mean Surface Temp. is : ",S_store[i])
        break
```
